```
import numpy as np
import matplotlib.pyplot as plt

# Вихідні дані для варіанта 28
x_values = np.array([-3, -1, 0, 2])
f_values = np.array([-16, 14, 5, -1])
given_points = np.array([-2, -1.5, -0.5, 1])

# Функція для обчислення інтерполяційного багаточлена Лагранжа
def lagrange_polynomial(x_values, f_values, x):
    n = len(x_values)
    result = 0
    for i in range(n):
        term = f_values[i]
        for j in range(n):
            if i != j:
                term *= (x - x_values[j]) / (x_values[i] - x_values[j])
        result += term
    return result

# Обчислюємо наближені значення функції у заданих точках
approx_values = [lagrange_polynomial(x_values, f_values, x) for x in given_points]

# Виведення результатів
for x, f_approx in zip(given_points, approx_values):
    print(f"f({x}) ≈ {f_approx:.4f}")

# Побудова графіка
x_plot = np.linspace(min(x_values) - 1, max(x_values) + 1, 100)
y_plot = [lagrange_polynomial(x_values, f_values, x) for x in x_plot]

plt.plot(x_plot, y_plot, label='Інтерполяційний багаточлен Лагранжа')
plt.scatter(x_values, f_values, color='red', label='Вихідні точки')
plt.scatter(given_points, approx_values, color='blue', label='Задані точки')
plt.legend()
plt.title("Інтерполяційний багаточлен Лагранжа")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
